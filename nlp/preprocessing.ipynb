```python
import requests
import pandas as pd

domain = "https://nos.nl"
# Define the endpoint URL
url = "/api/collectie/13959/items?limit=100&before=2519721"

# Set the initial parameters
params = {
    "limit": 100,
    "before": 2519721
}

def get_params_from_url(url):
    # Get the query string from the URL
    query_string = url.split("?")[1]
    
    # Split the query string into key-value pairs
    key_value_pairs = query_string.split("&")
    
    # Initialize an empty dictionary to store the parameters
    params = {}
    
    # Loop through the key-value pairs
    for key_value in key_value_pairs:
        # Split the key-value pair into key and value
        key, value = key_value.split("=")
        
        # Add the key-value pair to the dictionary
        params[key] = value
    
    return params

def get_url_without_params(url):
    # Split the URL into base URL and query string
    base_url, query_string = url.split("?")
    
    # Return the base URL
    return base_url

# Initialize an empty list to store the data
data = []

# Loop through the pages until there is no more data
while True:
    # Make the HTTP GET request
    params = get_params_from_url(url)
    base_url = get_url_without_params(url)

    print ("Getting data from URL:")
    print (domain + base_url)
    print (params)
    response = requests.get(domain + base_url, params=params)
    
    # Check if the request was successful
    if response.status_code == 200:
        # Parse the JSON response
        json_data = response.json()
        
        # Get the items from the response
        items = json_data.get("list", [])
        
        # Add the items to the data list
        data.extend(items)
        
        # Check if there are more pages
        if "nextLink" in json_data and json_data["nextLink"] is not None:
            # Update the parameters to get the next page
            url = json_data["nextLink"]
        else:
            # Break the loop if there are no more pages
            break
    else:
        # Print an error message if the request was not successful
        print("Error:", response.status_code)
        break

# Create a pandas dataframe from the collected data
nos_articles = pd.DataFrame(data)
nos_articles
```

```python
# define which files to process
summary_file = 'summary_20231220_articles'
```

```python
# load scraped data
df = pd.read_csv('../data/'+summary_file+'.csv')
df['date'] = pd.to_datetime(df['date'])
df = df[df['date'] > '2023-10-6']
```

```python
rx = r"\.(?!(?<=\d\.)\d)(?=\S)"

# fix some of the full stop formatting in the original files
df['text'] = df['text'].apply(ast.literal_eval)
df['text'] = df['text'].apply(lambda x: re.sub(rx, ". ", ' '.join([s for s in x])))
```

```python
#remove any null articles
df = df[df['date'].notnull()]
df['title'].fillna('', inplace=True)
```

```python
def generate_article_files(row):
    text = ""
    
    _id = row['id']
    title = row['title']
    date = row['date'].date()
    content = row['text']
    
    # Assemble as part of large text block
    text += " F-N-S-T-A-R-T. "
    text += str(_id)
    text += " F-N-E-N-D. "
    text += " D-A-T-E-S-T-A-R-T. "
    text += date.strftime(format="%Y-%m-%d")
    text += " D-A-T-E-E-N-D. "
    text += " T-I-T-L-E-S-T-A-R-T. "
    text += title
    text += " T-I-T-L-E-E-N-D. "
    text += " T-E-X-T-S-T-A-R-T. "
    text += content
    text += " T-E-X-T-E-N-D. "
    
    # Save current text chunk
    article_filename = "articles/article_" + str(_id) + ".txt"
    results_filename = "results/article_"+str(_id)+".txt.json"
    
    with open(article_filename, 'w') as f:
        f.write(text)
        
    #create dataframe which points to expected files
    row['article_file'] = article_filename
    row['results_file'] = results_filename
    
    return row

# generate article file in the format required for stanford nlp
df = df.apply(generate_article_files,axis=1)
```

```python
df.to_csv('./data/'+summary_file+'.csv', index=False)
```
